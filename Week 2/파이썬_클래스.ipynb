## class Human

````python
class Human():

  def __init__(self, email, address):
    self.email = email
    self.address = address

  def verify(self):
    if validate_email(self.email) == True:
      print(f"이메일 주소가 ({self.email}) 유효합니다!")
    else:
      print(f"이메일 주소가 ({self.email}) 유효하지 않습니다!")

  def send_mail(self):
    print(f"우편 메일이 {self.address} 로 보내집니다.")
````

````python
# 객체 "김중앙" 생성
                # 김중앙의 이메일                 # 김중앙의 주소
김중앙 = Human("middlekim@gmail.com", "서울특별시 성북구 동소문로 57")

# 김중앙의 이메일은 유효한 메일주소인가?
김중앙.verify()

# 김중앙이 입력한 메세지를 이메일로 보내기
김중앙.send_mail()
````

````python
# 객체 "이정보" 생성
                # 이정보의 이메일                 # 이정보의 주소
이정보 = Human("infolee@gmail.com", "경기도 성남시 분당구 성남대로 925번길")

# 이정보의 이메일은 유효한 메일주소인가?
이정보.verify()

# 이정보가 입력한 메세지를 이메일로 보내기
이정보.send_mail()
````
