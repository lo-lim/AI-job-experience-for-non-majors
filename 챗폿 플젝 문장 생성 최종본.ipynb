기본 문장에서 끝말과 필러 랜덤으로 형성 (약 450문장)

```python
sentense = "기능 알려줘."
endings = ["알려줘봐.", "알려줘라.", "알려주세요.", "부탁해.", "소개해줘.", "소개해봐.", "소개해주세요.", "궁금해.", "궁금해요.", "ㄱㄱ", "보여줘.", "보여주세요.", "알려줄래?", "보여줄래?"]
outputs = [re.sub("\w+\.$",x,sentense) for x in endings]
outputs

```

필러가 문법에 어긋나게 문장의 공백 사이에 들어가는 걸 방지하기 위해 소개해 줘>소개해줘, 소개해 봐>소개해봐, 소개해 주세요>소개해주세요 등으로 일부러 띄어쓰기 안 함. (추후 re.sub 함수로 문법에 맞게 수정할 예정)

봉구님 코드 참고

```python
fillers = ["근데","저기", "음", "흠", "있잖아", "어", "아", "저", "일단", "먼저", "우선"]
```

```python
sentense_list=[]  #만들어진 문장을 리스트 형식에 넣기 
```

```python

for i in outputs:      
    split_i = i.split(" ") #outputs을 하나 씩 i에 할당하는데 그 i를 공백을 기준으로 분리 
    if random.random()<0.1:   #확률로 처리 
        for num in range(len(split_i)):
            new_sent_i = split_i[:]
            filler = random.choice(fillers)
            new_sent_i.insert(num,filler)
        
            a = (" ".join(new_sent_i))  
        
            # b=re.sub("소개해주세요.", "소개해 주세요.", a)
            # b=re.sub("소개해봐.", "소개해 봐.", a)
            # b=re.sub("소개해줘.", "소개해 줘.", a)
            
            print(a) #fillers를 랜덤으로 넣은 문장들을 확인
            
            sentense_list.append(a)  #리스트 형식으로 되어 있는 setense_list에 위에 생성된 문장들을 하나 씩 추가 
     
```

```python
print(len(sentense_list))  #sentese_list에 있는 원소, 즉 문장들의 개수가 몇 개인지
new_list=set(sentense_list) #중복된 문장 제거 
print(len(new_list)) #중복을 제거하고 남은 문장들의 수 
```

```python
print(new_list)
```

```python
sentense_list2 = []
for k in new_list:
    sentense_list2.append(k)  #중복을 제거한 내용을 리스트 형태에 차례로 추가하기 
print(sentense_list2)
print(len(sentense_list2))
```

은주님 코드 참고

```python
first= ["", "너", "제일 편리한", "제일 많이 쓰이는", "사람들이 자주 쓰는", "사람들이 많이 쓰는", "가장 편리한", "모든", "챗봇", "자주 쓰는", "많이 쓰는"]
endings = ["", "알려줘봐.", "부탁해.", "좀 알려줘.", "좀 보여줘.", "어디가 좋아?", "알려줄래?", "보여줘.", "알려줘.", "뭐 있어?", "으로 뭐 할 수 있어?", "보고 싶어.", "보여주세요.", "소개해 줘.", "소개해 봐.", "궁금해.", "궁금해요.", "ㄱㄱ"]
function= ["기능", "능력"]


result = set()    #중복되어 있는 문장을 제거 한 채로 저장 

sample= "너의 기능 알려줘."
result.add(sample)
n = 190

while len(result) < n:   #문장의 수가 317개까지만 반복해서 생성해라 
    sent = re.sub('너의', random.choice(first), sample)
    print(sent)
    sent = re.sub('알려줘[.]', random.choice(endings), sent)
    print(sent)
  
    sent = sent.strip()
    sent = re.sub('\s{2}', ' ', sent)
    result.add(sent)

open('result_1.txt', 'w').write('\n'.join(result))
```

```python
sentense_list3 = []
for k in result:
    sentense_list3.append(k)  
print(sentense_list3)

df_2=pd.DataFrame({"문장": sentense_list3})
df_2.tail()
```

```python
s= "너 뭐 할수 있는지 알려줘."
e= ["말해.", "말해줘.", "알려줘봐.", "알려주세요.", "ㄱㄱ.", "말해볼래?", "소개해 줘.", "소개해 주세요.", "궁금해.", "궁금해요."]
out=[re.sub("\w+\.$",x,s) for x in e]
out
```
