# Routing & 변수사용
기본적인 홈페이지 외에 create용 페이지를 만들어 'Create Time!'을 출력시키고 create로 라우팅

create페이지 말고 readid/[id number] 주소로 새로운 페이지를 만든 후 주소에 적혀있는 id number를 이용하여 'Welcome number[id number]!'의 문구를 출력시키기

```python
app = Flask(__name__)
run_with_ngrok(app)

@app.route('/')
def home():
  return f"<h1>this is homepage</h1>"


@app.route('/create/')
def create():
  return f"<h1>Create Time!</h1>"



app.run()
```
