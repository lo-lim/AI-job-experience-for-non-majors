### 공백의 유형
* 자연어 처리에서 흔히 기본적으로 처리하는 공백의 유형은 2가지입니다.
* double space
* 앞 뒤 공백

```python
double_space = re.compile('  +')
space_head = re.compile('^\s.+')
space_tail = re.compile('.+\s$')

def find_spaces(dataframe):
  for row in dataframe.itertuples():
    text = row[1]
    if re.findall(double_space, text):
      print(f"중복 공백 : {text}")
    if re.findall(space_head, text):
      print(f"앞 공백 : {text}")
    if re.findall(space_tail, text):
      print(f"뒷 공백 : {text}")

find_spaces(new_df)
```

###### 글자수 필터링
- 적정 글자수 최소 2글자 최대 20글자
- 참고 : [pandas.Series.apply](https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html)
* apply method: 판다스의 series 자료형에, 원하는 함수를 자유롭게 적용할 수 있는 method.

```python
a = pd.Series([1,2,3,4,5])
print(a)
```

```python
a = a.apply(lambda x:x**2)
print(a)
```

```python
a = a.apply(lambda x : x > 15)
print(a)
```

```python
len_check = new_df["utt"].apply(lambda x: len(x) >= 5) & new_df["utt"].apply(lambda x: len(x) <= 20)
print(len_check)
```
