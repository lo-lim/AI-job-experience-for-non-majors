## 정규표현 복습
* import re
* re.search, re.sub, re.findall
* re.search("패턴", text)
* re.sub("패턴", "바꿀 결과", text)
* re.findall("패턴", text)

```python
text = "안녕하세요"

# "하" 찾아보기.
re.search("하", text)   # 위에 "패턴"에 찾고자 하는 문자를 입력하면 됨 
```

```python
# "세"를 "안녕"으로 바꿔보기
text = re.sub("세", "안녕", text)
print(text)
```

```python
# "안녕" 전부 찾아보기  #findall은 존재하는 문자를 있는 개수만큼 다 출력 
re.findall("안녕", text)
```

### 메타 문자. 빈칸 채워보기!
* 패턴의 자리에 사용하는 문자
* 숫자: [0-9], \d
* 문자: \w, 가나다라마바사 abcdefg, 즉 숫자 혹은 한글 
* 공백: [ ], \s
* 한글 글자: [가-힣]
* 한글 자모: [ㄱ-ㅎ]
* 영어 대문자: [A-Z]
* 영어 소문자: [a-z]
* 숫자가 아닌 문자: [^0-9], [^\d]숫자가 안 잡힙. 즉 ㄱㄴㄷㄹㅁㅂㅅ 와 같은 한글
* 모든 문자(\n 제외): 

#### 반복 메타 문자
* ?: 0-1  (0부터1까지의 횟수)
* *: 0-x  (0이상)
* +: 1-x  (1이상)


### 문제
* 다양한 역명을 전부 "안암역"으로 바꿔보기

```python
text = '''
이대역 근처 카페 좀 알려줘.
이대역 맛집 좀 알려줄 수 있니.
여의도역 주변 식당 좀 알려줘.
여의도역 유명한 맛집 좀 알려줄 수 있니.
신촌역 괜찮은 식당 어떤 거 있니?
신촌역 카페 어디 있어?
안양역 주변 식당 어떤 데가 좋아?
안양역 근처에 카페 어떤 거 있는지 알려줄 수 있니?
홍대역 맛집 어디 있어?
홍대역 카페 어디가 괜찮아?
동대문역 주변 음식점 좀 알려줘.
동대문역 주변 카페 좀 알려주라.
'''
```

```python
text= re.sub(".+역", "안암역", text)
print(text)
```

### 문제
* 아래 텍스트에 대해서 "이대역"을 "안암역"으로 바꿔보기.
* 이대역을 직접 쓰지 않고, 메타 문자를 활용해야 함!

```python
text = "근처 가장 가까운 이대역 맛집 좀 알려줘."
```

```python
# text= re.sub("[^ ]+역", "안암역", text)
text=re.sub("[가-힣]+역", "안암역", text)
print(text)
```
