## 필러 삽입 및 어순 섞기
* 구현한 코드는 드립니다. 좀 어려울 수 있습니다. 관심 있는 사람은 참고하세요.


* <font color=blue>핵심 아이디어:</font> rand_num = random.randint(0, len(split_sent))  --> 랜덤으로 필러 삽입 / 어순 섞기 지점 설정
* SPACE_EXISTING = True  --> "분위기 좋은"과 같은 경우를 분리하지 않기 위한 변수

```python
import re
import random

accomodation = ["호텔", "잘 곳", "게스트하우스", "독채", "독채펜션", "펜션", "풀빌라", "모텔", "머무를 곳"]
pretty = ["", "분위기 좋은", "편안한", "마당 있는", "가족끼리 갈만한", "괜찮은", "감성", "오션뷰", "바다뷰", "바다 보이는", "한달살기"]
jeju = ["", "제주", "애월", "제주 서부", "제주 동부", "서귀포", "하귀", "제주공항", "제주공항 근처", "성산일출봉 근처", "우도"]
endings = ["", "추천", "어디가 좋니?", "좀 추천해줘.", "추천 좀", "어디가 좋아?", "알려줄래?", "어디가 괜찮아?", "추천해줘.", "어디가 예뻐?", "어디에 있어?", "어디가 괜찮니?", "알려줘.", "어디가 좋을까?"]
filers = ["음", "뭐지", "그", "흠", "어", "저"]
result = set() # 랜덤하게 생성된 문장을 저장할 집합(for 중복 제거)

sample = "제주도 예쁜 숙소 추천해줘."
result.add(sample)
n = 715  # 생성할 문장 개수
SPACE_EXISTING = False

while len(result) < n:
    rand_list = [random.choice(jeju), random.choice(pretty), random.choice(accomodation), random.choice(endings)]
    
    sent = re.sub('제주도', rand_list[0], sample)
    sent = re.sub('예쁜', rand_list[1], sent)
    sent = re.sub('숙소', rand_list[2], sent)
    sent = re.sub('추천해줘[.]', rand_list[3], sent)


    # 필러 추가 (if random.random() <= 뒤의 숫자로 확률 조정 가능)
    if random.random() <= 0.1:     #10프로의 확률이라는 논리가 생김 
        split_sent = sent.split()
        rand_num = random.randint(0, len(split_sent))      #필러 추가할 인덱스(자리)를 랜덤 선택
        # print(rand_num)
        filler = random.choice(filers)
        split_sent.insert(rand_num, filler)
        sent = ' '.join(split_sent)
        # print(sent)
            


    # 순서 섞기
    for word in rand_list:
        if re.search(' ', word):     # 선택된 단어에 공백이 존재하는지 판단 ex) "잘 곳", "분위기 좋은"
            SPACE_EXISTING = True    
            # 공백이 존재하면, SPACE_EXISTING = True / 그렇지 않으면 그대로 SPACE_EXISTING = False
            
    # 순서를 섞을 조건: "not SPACE_EXISTING": 즉, 선택된 단어에 공백이 없을 때
    if not SPACE_EXISTING and len(sent.split()) > 2:
        
        split_sent = sent.split()
        rand_num = random.randint(1, len(split_sent) - 1)      # 순서 섞을 인덱스(자리)를 랜덤 선택
        first = sent.split()[:rand_num]
        second = sent.split()[rand_num:]
        sent = ' '.join(second+first)
        
        ''' 종결 어미 처리. 순서 섞는 과정에서 문장의 중간으로 이동한 종결 어미를 문장의 끝으로 보낸다.
        그리고 그 자리에 콤마(,) 삽입
        ex) "제주도 호텔 추천해줘!" --> 추천해줘! 제주도 호텔 --> 추천해줘, 제주도 호텔!'''
        if re.search('[.!?,]', sent):
            punc = re.search('[.!?,]', sent).group(0)
            sent = re.sub('['+ punc+']', ',', sent)
            sent += punc
        print(sent)

        

    sent = sent.strip()  # 문장 앞뒤 공백 제거
    sent = re.sub('\s{2}', ' ', sent)  # 스페이스 2번 쓰인 것을 1번으로
    result.add(sent)  # 집합의 원소를 더하는 메서드 add
    SPACE_EXISTING = False

open('result.txt', 'w').write('\n'.join(result))
```

#### for 구문을 통해 모든 자리에 필러를 다 삽입했음.
* 그것을 딱히 확률로 제어하지도 않았음.
* 모든 자리에 골고루 필러가 들어가는 데이터 형성.
* 또한 "test_list.append(a)  #이 부분을 반복해서 test_list에 여러 filler와 endings가 들어간 문장 반복생성"라고 하셨는데,
* 아까도 말했다시피 코드의 핵심은, 반복을 줄이는 것!
* "여러 번 반복 실행"을 직접 수동으로 할 거면 코드를 쓸 이유가 없다.
* "여러 번 반복 실행"을 for 구문으로 처리해야 함!

```python
for i in outputs:#endings를 수정한 리스트 원소 지정, 수만큼 반복
    split_i = i.split(" ")
    for num in range(len(split_i)): #split한 문장 사이사이 필러 넣을 자리
        new_sent_i = split_i[:]
        filler = random.choice(fillers)
        new_sent_i.insert(num,filler)
        
        a = (" ".join(new_sent_i))  #
        
        print(a) #만들어진 문장 확인
        test_list.append(a)  #이 부분을 반복해서 test_list에 여러 filler와 endings가 들어간 문장 반복생성
```

```python
print(len(test_list)) #반복 시행으로 생성한 문장이 들어간 리스트의 길이

set_list = set(test_list) #중복 제거
print(len(set_list)) #중복 제거한 길이
```

### 수정된 코드
* for filler in fillers:를 추가하여, 모든 자리에 모든 filer를 다 넣었다.

```python
# 수정 후
sentense = "카페 추천 해줘."
endings = ["해주라.","해봐.","해라.","부탁해.","해.", "해줄래?","ㄱㄱ","해볼래?.","해봐봐.","하라니까.","해보라니까.","부탁해.","해볼래?"]
outputs = [re.sub("\w+\.$",x,sentense) for x in endings]
print(outputs)

test_list = []
fillers = ["근데","있잖아","저기","음","흠","그","아","저","그런데","일단","그니까","그러니까","저기","좀"]
     
for i in outputs:
    split_i = i.split(" ")
    for num in range(len(split_i)):
        
        # 수정 부분
        for filler in fillers:
          new_sent_i = split_i[:]
          new_sent_i.insert(num,filler)
        #필러를 또 for 구문에 넣어서 수동으로 개수를 생성하는 걸 피할 수 있음

          
          a = (" ".join(new_sent_i))
        
          print(a) #만들어진 문장 확인
          test_list.append(a)  #이 부분을 반복해서 test_list에 여러 filler와 endings가 들어간 문장 반복생성

print(len(test_list)) #반복 시행으로 생성한 문장이 들어간 리스트의 길이

set_list = set(test_list) #중복 제거
print(len(set_list)) #중복 제거한 길이
```

## indent로 인한 time complexity 문제
* open() , write() 엄청 자주 쓰는데, 조심해야 합니다!
* open: 파일 생성
* write: open된 파일에 내용 작성

```python
import random
import re
import pandas as pd

sentence = "제주도 관광지 추천 해줘."
endings = ["해주라.",'해봐.','해주세요.','좀 해줘.','해.','해봐바.','해라.','해줭.','해줄랭.','ㄱㄱ.','해줘봐라.','해볼래.']
fillers = ['흠','음','아','어']
location = ['제주시','서귀포시','제주','서귀포','제주섬','제주쪽','서귀포쪽','섬 위쪽','섬 아래쪽','제주도 아래쪽','제주도 위쪽','제주도 동쪽','제주도 서쪽','제주도 북쪽','제주도 남쪽']
data=[]


outputs=[re.sub("\w+\.$",x,sentence) for x in endings]
for x in outputs:
    outputs1 = [re.sub('제주도',y,x) for y in location]
    for x in outputs1:
        split_sent=x.split()
        for idx in range(len(split_sent)):
            new_sent = split_sent[:]
            filler = random.choice(fillers)
            new_sent.insert(idx,filler)
            result=(' '.join(new_sent))
            data.append(result)
            # 여기서 파일을 저장해버리면, 파일 덮어쓰기 저장을 for 구문 반복 횟수만큼 반복하는 것!
            open('result4.txt', 'w').write('\n'.join(data))   #여기서 내부에 적으면 안 되고 들여쓰기를 안 하고 적어야 함 
```

## 문장이 과하게 많이 출력됩니다. 이유는?
* print(sent)를 4번이나 써놨고, 중복이 있기 때문!
* random.choice()를 1번 할 때마다 print(sent)를 한 번씩 써놨는데, 
* 예를 들어 weather의 경우 원소가 5개밖에 없는데 random.choice(weather)를 수 백 번 이상 한다면 당연히 중복이 있을 것!
* 그러한 중복의 경우들이 전부 출력되므로 문장이 과하게 많이 출력될 수밖에 없다.


* 물론 최종 결과물은 파이썬의 set()을 거치기 때문에 그러한 중복들이 전부 사라진다.

```python
import re
import random

weather = ["날씨", "기상 정보", "기온", "날씨 상태", "날씨 상황"]
date = ["오늘", "현재", "이번주", "지금", "이번에", "내일", "내일 모레", "이번주", "다음주", "저번주"]
jeju = ["제주", "애월", "제주 서부", "제주 동부", "서귀포", "제주공항", "제주공항 근처", "성산일출봉 근처", "우도"]
endings = ["알려줘.", "어떤지 알려줘.", "궁금해.", "말해줘.", "궁금하다.", "궁금해요.", "알고 싶어.", "알고 싶어요.", "알려 줄래?", "어때요?",  "어떤가요?", "어떠니?"]

result = set()

sample = "오늘 제주도 날씨 알려줘."
result.add(sample)
n = 750

while len(result) < n:
    sent = re.sub('날씨', random.choice(weather), sample)
    print(sent)
    sent = re.sub('오늘', random.choice(date), sent)
    print(sent)
    sent = re.sub('제주도', random.choice(jeju), sent)
    print(sent)
    sent = re.sub('알려줘[.]', random.choice(endings), sent)
    print(sent)

    sent = sent.strip()
    sent = re.sub('\s{2}', ' ', sent)
    result.add(sent)

open('result.txt', 'w').write('\n'.join(result))
```
