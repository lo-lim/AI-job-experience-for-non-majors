```python
# ì¤‘ë³µ ê³µë°±, ì•, ë’¤ ê³µë°± ì²´í¬
# <> ì±„ìš°ê¸°
double_space = re.compile('<>')
space_head = re.compile('<>')
space_tail = re.compile('<>')

# ê³µë°± ë¬¸ì œìˆëŠ” utt ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def find_spaces(dataframe):
  for row in dataframe.itertuples():
    text = row[1]
    if re.findall(double_space, text):
      print(f"ì¤‘ë³µ ê³µë°± : {text}")
    if re.findall(space_head, text):
      print(f"ì• ê³µë°± : {text}")
    if re.findall(space_tail, text):
      print(f"ë’· ê³µë°± : {text}")

find_spaces(data_df)
```
