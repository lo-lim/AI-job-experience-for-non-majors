## train / validation / test split

*데이터를 TRAIN SET---> 0.8/ VAL SET---> 0.1/ TEST SET---> 0.1
* 모델을 훈련할 때는 거의 100프로 train / val / test split이 필요하다고 보시면 됩니다.
* 흔히 모델을 훈련하는 데이터는 train data입니다. 그와 별도로 val, test data가 왜 필요할까요?
* "모델 성능 평가"를 위해 필요합니다. train data 바탕으로 모델을 훈련한 뒤,
* val / test 데이터를 통해 성능을 평가 및 개선합니다.
* "train 때 전혀 못 본 완전히 낯선 데이터"로 성능을 평가하기 위해 분리하는 것입니다.

*성능 평가를 위한 데이터의 조건: "unseen" 
---> 낯선 데이터 train validation test. --> 

*validation/test
val: train의 결과를 향상시키기 위해 성능을 보조적으로 평가
test: 최종적으로 모델 성능 평가를 위한 data set

```python
import pandas as pd

# 경로를 적절히 설정해주세요
file_path =  '/content/drive/MyDrive/Colab Notebooks/챗폿 프로젝트/4team_0917/processed.csv'
data_df = pd.read_csv(file_path, index_col=0, header=0)
data_df
```

### 판다스 sample method --> 랜덤 샘플 선택
* pd.DataFrame.sample()을 통해 랜덤하게 데이터를 분리할 수 있습니다.
* frac=비율  -->을 통해 선택하는 비율을 설정할 수 있습니다.
* random_state의 역할은 뭘까요?

### 문제!
* random_state= 뒤에 원하는 숫자 값을 아무 거나 넣어보면서 여러 번 실행해 봅시다.

```python
df_train = data_df.sample(frac=0.8, random_state=1)
print(df_train)
print()
print(f"length of train df: {len(df_train)}")

#전체 24개의 개수에서 19개 출력 
#랜덤하게 분리하기에 인덱스가 순서대로가 아니라 랜덤으로 뽑힌 상태 
#frac=0.8은 전체 개수에서 80%에 해당하는 것. 그래서 19개만 추출된 것 
```

```python
df_train = data_df.sample(frac=0.8, random_state=3)
print(df_train)
```

* random_state=k에서 k의 수에 따라서 그 값만큼 고정되어서 나온다는 것. k를 무엇으로 설정하냐에 따라서 랜덤을 우리가 컨드롤할 수 있음
